```python
import pandas as pd

df = pd.read_csv("./sns_recommand.csv", sep="|")
df
```

```python
df['result'] = df['contents'] + " " + df['comments']
```

```python
def combine_2rd_columns(col_1, col_2):
    result = col_1
    if not pd.isna(col_2):
        result += " " + str(col_2)
    return result

df["result"] = df.apply(lambda x: combine_2rd_columns(x['contents'], x['comments']), axis=1)
```
